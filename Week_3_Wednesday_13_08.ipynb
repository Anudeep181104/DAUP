### Task 1: Phishing Email Detection Simulation (Scatter Plot for Email Data)

This code will create a scatter plot showing the frequency of "account" and "login" words in spam and non-spam emails. Make sure you have the `emails.csv` file in your Colab environment (e.g., in the `/content/` directory) for this code to work.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset - make sure 'emails.csv' is in the correct path
try:
    df_emails = pd.read_csv("/content/emails.csv")

    # Drop non-numerical columns (keeping only word counts + spam labels)
    # Assuming 'Prediction' column contains the spam labels (0 or 1)
    if 'Prediction' in df_emails.columns:
        df_numeric_emails = df_emails.select_dtypes(include=['int64', 'float64'])
        df_numeric_emails['label'] = df_emails['Prediction']

        # Create scatter plot
        plt.figure(figsize=(10, 7))

        # Separate spam and non-spam emails
        spam_emails = df_numeric_emails[df_numeric_emails['label'] == 1]
        non_spam_emails = df_numeric_emails[df_numeric_emails['label'] == 0]

        # Scatter plot for spam emails (red stars)
        if 'account' in spam_emails.columns and 'login' in spam_emails.columns:
            plt.scatter(spam_emails['account'], spam_emails['login'], color='red', marker='*', label='Spam Emails', alpha=0.6)
        else:
             print("Warning: 'account' or 'login' columns not found in spam emails data.")


        # Scatter plot for non-spam emails (blue circles)
        if 'account' in non_spam_emails.columns and 'login' in non_spam_emails.columns:
             plt.scatter(non_spam_emails['account'], non_spam_emails['login'], color='blue', marker='o', label='Non-Spam Emails', alpha=0.6)
        else:
             print("Warning: 'account' or 'login' columns not found in non-spam emails data.")


        # Add title, labels, and legend
        plt.title("Account & Login Word Frequency in Spam vs Non-Spam Emails")
        plt.xlabel("Count of 'account'")
        plt.ylabel("Count of 'login'")
        plt.legend()
        plt.grid(True)
        plt.show()

    else:
        print("Error: 'Prediction' column not found in the dataset.")

except FileNotFoundError:
    print("Error: emails.csv not found. Please upload the file to /content/.")
except Exception as e:
    print(f"An error occurred: {e}")
```

### Task 2: Hours Studied vs Test Scores (Scatter Plot)

This code will create a scatter plot showing the relationship between hours studied and test scores, using the example data.

```python
import matplotlib.pyplot as plt

# Example data for hours studied and test scores
hours_studied = [2, 3, 4, 5, 6, 7, 8, 9, 10]
test_scores = [40, 50, 55, 65, 70, 75, 80, 85, 95]

# Create scatter plot
plt.figure(figsize=(8, 6))

# Plot data points with custom colors and markers
high_scores_plotted = False
low_medium_scores_plotted = False

for i in range(len(hours_studied)):
    if test_scores[i] >= 75:
        # High scores in red with triangles
        label = 'High Score' if not high_scores_plotted else ""
        plt.scatter(hours_studied[i], test_scores[i], color='red', marker='^', label=label)
        high_scores_plotted = True
    else:
        # Low/Medium scores in blue with circles
        label = 'Low/Medium Score' if not low_medium_scores_plotted else ""
        plt.scatter(hours_studied[i], test_scores[i], color='blue', marker='o', label=label)
        low_medium_scores_plotted = True


# Add title, labels, and legend
plt.title("Study Hours vs Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")
plt.legend()
plt.grid(True)
plt.show()
```
